## Part(a) :  Mean Squared Error (MSE)
Write a Python function that computes the Mean Squared Error loss given:
- True labels y
- Predicted values y_hat

```python
%%writefile mse_loss.py
import numpy as np
def mse_loss(y,y_hat) -> np.ndarray:
    """
    Python function that computes the Mean Squared Error loss given:
    - True labels y
    - Predicted values y_hat
    """
    return np.mean(y-y_hat)**2
```

```python
from mse_loss import mse_loss
print("="*70)
print('example equation : y = x+1')
print("="*70)

#keep the reproducibility
np.random.seed(42)

x = np.random.randint(0,50,50)
print("Input Value  :\n" , x)
print()

y = x+1
print("True Value : \n" , y)
print()

mu, sigma = 0, 1 # mean and standard deviation
noise = np.random.normal(mu,sigma,50)
# print("Noise Value : \n" , noise)

y_hat = np.round(x+noise,decimals=2)
print("predicted value:\n",y_hat)
print()

print("Mean Squared Error :", np.round(mse_loss(y,y_hat),decimals=2))
```

## Part(b) : Binary Cross-Entropy Loss
Write a Python function that computes the Binary Cross-Entropy loss for a binary
classification task given:
- True labels y ∈ {0,1}
- Predicted probabilities  y_hat ∈ (0,1)

```python
%%writefile binary_cross_entropy_loss.py
import numpy as np
def binary_cross_entropy_loss(y,y_hat) -> np.ndarray:
    """
    Python function that computes the Binary Cross-Entropy loss for a binary
    classification task given:
    - True labels y ∈ {0,1}
    - Predicted probabilities y_hat ∈ (0,1) Performs clipping to avoid log(0).
    """
    epsilon = 1e-15  # A very small number to prevent log(0)
    y_hat = np.clip(y_hat, epsilon, 1 - epsilon)
    return -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
```

```python
from binary_cross_entropy_loss import binary_cross_entropy_loss
print("="*70)
print('example equation : predicted_class for feature (x) : 1/1+e^-x')
print("="*70)

#keep the reproducibility
np.random.seed(42)

x = np.random.randint(-10,10,50)
mu = 0
sigma = 1
noise = np.random.normal(mu,sigma,50)
x = np.round(x+noise,decimals=2)
print("Input Feature  :\n" , x)
print()

y_true_class = np.random.randint(0,2,50)
print("True Class :\n" , y_true_class)
print()

y_predicted_probability = np.round(1/(1+np.exp(-x)),decimals=3)
print("Predicte Probability :\n" , y_predicted_probability)
print()

print("Binary Cross  Entropy :", np.round(binary_cross_entropy_loss(y_true_class,y_predicted_probability),decimals=2))
```
